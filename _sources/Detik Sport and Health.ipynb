# Tugas 4 Crawling Berita Detik (Sport, Health, Travel)

## Crawling Data

- Impor Library:

import requests: Digunakan untuk mengirim permintaan HTTP ke situs web dan mengunduh halaman web.
from bs4 import BeautifulSoup: Digunakan untuk mem-parsing HTML halaman web.
import nltk: Library pemrosesan bahasa alami (Natural Language Processing).
import csv: Digunakan untuk menulis data ke file CSV.

- Fungsi get_article_text:

Fungsi ini menerima URL artikel, mengunduh halaman, dan mengekstrak teks dari div dengan kelas CSS 'detail__body-text'. Fungsi ini mengembalikan teks artikel.

- Inisialisasi Variabel dan Membuka File CSV:

base_url: URL dasar untuk indeks detiksport.
csv_file: Membuka file CSV untuk menulis data, dengan header 'Judul Artikel' dan 'Isi Artikel'.

- Iterasi Melalui 20 Halaman:

Menggunakan loop untuk mengakses 20 halaman detiksport/indeks.
Membuat URL untuk setiap halaman dan mengirimkan permintaan HTTP.

- Parsing Halaman Web dan Ekstraksi Tautan Artikel:

Menggunakan BeautifulSoup untuk mem-parsing HTML halaman web.
Menggunakan CSS class 'media__link' untuk menemukan tautan artikel.
Iterasi melalui setiap tautan artikel.

- Dapatkan Tautan dan Judul Artikel:

Mendapatkan URL dan judul artikel dari setiap elemen tautan artikel.

- Dapatkan Teks Artikel dan Tulis ke File CSV:

Memanggil fungsi get_article_text untuk mendapatkan teks dari setiap artikel.
Menulis judul artikel dan teks artikel ke dalam file CSV.

- Menutup File CSV:

Menutup file CSV setelah selesai menulis data.

```python
import requests
from bs4 import BeautifulSoup
import nltk
import csv

# Fungsi untuk mendapatkan teks dari sebuah artikel
def get_article_text(article_url):
    article_response = requests.get(article_url)
    if article_response.status_code == 200:
        article_soup = BeautifulSoup(article_response.text, 'html.parser')
        article_content = article_soup.find('div', class_='detail__body-text')
        if article_content:
            return article_content.get_text(strip=True)
    return ''

# Base URL for detiksport/indeks
base_url = "https://sport.detik.com/indeks/"

# Membuka file CSV untuk menulis data
csv_file = open('detiksport_articles.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Judul Artikel', 'Isi Artikel'])

# Iterasi melalui 20 halaman
for page_number in range(1, 21):
    # Buat URL untuk halaman tertentu
    url = f"{base_url}{page_number}"
    response = requests.get(url)
    html = response.text

    # Parsing halaman web menggunakan BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    # Ekstraksi tautan artikel dan judul artikel
    article_elements = soup.find_all('a', class_='media__link')

    # Iterasi melalui setiap tautan artikel
    for article_element in article_elements:
        # Dapatkan tautan artikel dan judul artikel
        article_url = article_element['href']
        article_title = article_element.get_text(strip=True)

        # Dapatkan teks dari artikel menggunakan fungsi yang telah dibuat
        article_text = get_article_text(article_url)

        # Tulis judul artikel dan isi artikel ke file CSV
        csv_writer.writerow([article_title, article_text])

# Tutup file CSV
csv_file.close()
```
